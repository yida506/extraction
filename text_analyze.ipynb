```python
def new_df(df):
    parsedata = {}
    for i in df.iterrows():
        if i[1]['name'] in parsedata:
            parsedata[i[1]['name']].append(i[1]['target'])
        else:
            parsedata[i[1]['name']] = [i[1]['target']]
    return parsedata
```
